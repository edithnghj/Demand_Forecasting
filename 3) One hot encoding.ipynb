```python
import pandas as pd
```

```python
df = pd.read_csv('combined.csv')
df
```

### Dummify center_id

```python
encoded_center_id = pd.get_dummies(df['center_id'], prefix='center_id')
encoded_center_id
```

### Dummify meal_id

```python
encoded_meal_id = pd.get_dummies(df['meal_id'], prefix = 'meal_id')
encoded_meal_id
```

### Dummify city_code

```python
encoded_city_code = pd.get_dummies(df['city_code'], prefix='city_code')
encoded_city_code
```

### Dummify region code

```python
encoded_region_code = pd.get_dummies(df['region_code'], prefix='city_code')
encoded_region_code
```

### Dummify center_type

```python
encoded_center_type = pd.get_dummies(df['center_type'])
encoded_center_type
```

### Dummify category

```python
encoded_category = pd.get_dummies(df['category'])
encoded_category
```

### Dummify cuisine

```python
encoded_cuisine = pd.get_dummies(df['cuisine'])
encoded_cuisine
```

```python
df = pd.concat([df, encoded_center_id, encoded_meal_id, encoded_city_code, encoded_region_code, encoded_center_type, encoded_category, encoded_cuisine], axis = 1)
df
```

```python
drop_cols = ['id', 'center_id', 'meal_id', 'city_code', 'region_code', 'center_type', 'category', 'cuisine']
df = df.drop(drop_cols, axis=1)
df
```

```python
df.to_csv('combined_2.csv', index=False)
```
