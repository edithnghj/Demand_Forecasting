## Loading Data Sets

```python
import pandas as pd
```

```python
orders = pd.read_csv('data.csv')
orders.head()
```

```python
meals = pd.read_csv('meal_info.csv')
meals.head()
```

```python
centers = pd.read_csv('fulfilment_center_info.csv')
centers.head()
```

## Check for missing values

```python
orders.isnull().sum()
```

```python
meals.isnull().sum()
```

```python
centers.isnull().sum()
```

## Merging the Data Sets

```python
df = pd.merge(orders, centers, on='center_id')
df
```

```python
df = pd.merge(df, meals, on='meal_id')
df
```

```python
df.to_csv('combined.csv', index = False)
```
