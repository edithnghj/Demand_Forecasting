```python
import pandas as pd
```

```python
df = pd.read_csv('combined_2.csv')
df
```

```python
dX = df.drop(['num_orders'], axis=1)
dy = df['num_orders']

X = dX.values
y = dy.values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

### Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)

lr_rmse = np.sqrt(mean_squared_error(y_test, lr_pred))
print('RMSE for linear regression is:', lr_rmse)

plt.scatter(y_test, lr_pred)
plt.show()
```

### Decision Tree Regressor

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
dtr_pred = dtr.predict(X_test)

dtr_rmse = np.sqrt(mean_squared_error(y_test, dtr_pred))
print('RMSE for Decision Tree Regressor is:', dtr_rmse)

plt.scatter(y_test, dtr_pred)
plt.show()
```

### Random Forest Regressor

```python
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
rfr_pred = rfr.predict(X_test)

rfr_rmse = np.sqrt(mean_squared_error(y_test, rfr_pred))
print('RMSE for Random Forest Regressor is:', rfr_rmse)

plt.scatter(y_test, rfr_pred)
plt.show()
```

### Predict the demand for the next 10 weeks (Weeks: 146-155) for the center-meal combinations in the test set 


```python
week_135 = df[df['week']<=135]
week_135
```

```python
week_145 = df[(df['week']>135) & (df['week']<=145)]
week_145
```

```python
X_train = week_135.drop(['num_orders'], axis=1)
y_train = week_135['num_orders']

X_test = week_145.drop(['num_orders'], axis=1)
y_test = week_145['num_orders']
```

```python
model_rfr = RandomForestRegressor()
model_rfr.fit(X_train, y_train)
model_rfr_pred = model_rfr.predict(X_test)

model_rfr_rmse = np.sqrt(mean_squared_error(y_test, model_rfr_pred))
print('RMSE for Random Forest Regressor is:', model_rfr_rmse)

plt.scatter(y_test, model_rfr_pred)
plt.show()
```

```python
week_145['prediction'] = model_rfr_pred
week_145
```

```python
week_145[['week', 'num_orders', 'prediction']].groupby(week_145['week']).agg({'num_orders':'sum', 'prediction':'sum'})
```

```python
week_135[['week', 'num_orders']].groupby(week_135['week']).agg({'num_orders':'sum'})
```

```python
import seaborn as sns
plt.figure(figsize=(16,8))
sns.lineplot(x='week', y = 'num_orders', data=week_135, ci=None)
sns.lineplot(x='week', y = 'num_orders', data=week_145, ci=None, label='Actual')
sns.lineplot(x='week', y = 'prediction', data=week_145, ci=None, label='Predicted')
plt.legend()
plt.show()
```
